```python
from keras.models import Sequential 
from keras.layers import Dense
from keras.layers import MaxPooling2D
from keras.layers import Convolution2D
import cv2
import matplotlib.pyplot as plt
import numpy as np
```

```python
from keras.layers import Flatten
```

```python
#Initialising the CNN model
```

```python
classifier = Sequential ()

#Making the first convolved layers
classifier.add (Convolution2D (32, 3, 3, input_shape = (64, 64, 3), activation = "relu"))

#Making the first pooled layers
classifier.add (MaxPooling2D (pool_size = (2, 2)))

#Flattening all the pooled layers and passing them as the input nodes in the ANN architecture
classifier.add (Flatten ())

#Adding the ANN layers in the CNN
classifier.add (Dense (output_dim = 128, activation = "relu", init = "uniform"))
classifier.add (Dense (output_dim = 1, activation = "sigmoid", init = "uniform"))

#Compiling the CNN
classifier.compile (loss = 'binary_crossentropy', metrics = ['accuracy'], optimizer = 'adam')
```

```python
#Fitting the CNN to the images
from keras.preprocessing.image import ImageDataGenerator
```

```python
train_datagen = ImageDataGenerator (
    rescale = 1. / 255,
    shear_range = 0.2,
    zoom_range = 0.2,
    horizontal_flip = True
)
```

```python
test_datagen = ImageDataGenerator (
    rescale = 1. / 255
)
```

```python
training_set = train_datagen.flow_from_directory (
    'Convolutional_Neural_Networks/dataset/training_set',
    target_size = (64, 64),
    batch_size = 32,
    class_mode = 'binary'
)
```

```python
test_set = test_datagen.flow_from_directory (
    'Convolutional_Neural_Networks/dataset/test_set',
    target_size = (64, 64),
    class_mode = 'binary'
)
```

```python
classifier.fit_generator (
    training_set,
    samples_per_epoch = 8000,
    epochs = 25,
    validation_data = test_set,
    nb_val_samples = 2000
)
```
